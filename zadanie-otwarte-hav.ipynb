### Olympic swiming

Opis datasetu:

1. `Location`: Miejsce, w którym odbyły się igrzyska olimpijskie.
2. `Hosting City`: Miasto, które było gospodarzem danej edycji igrzysk olimpijskich.
3. `Year`: Rok, w którym odbyły się igrzyska olimpijskie.
4. `Distance`: Odległość w metrach, na jaką odbywały się konkurencje pływackie. W przypadku sztafet, informacja zawiera liczbę pływaków i dystans pływany przez jednego z nich (np. "4x200" oznacza sztafetę 4x200 metrów).
5. `Stroke`: Styl pływacki, którym odbywana była konkurencja. Może to być: "Backstroke" (plecy), "Breaststroke" (klasyczny), "Butterfly" (motyl), "Freestyle" (dowolny), "Medley" (zmienne) lub "Individual Medley" (indywidualne zmienne).
6. `Relay?`: Wartość określająca typ konkurencji. 0 oznacza konkurencje indywidualne, a 1 oznacza konkurencje sztafetowe (relay).
7. `Gender`: Płeć zawodników uczestniczących w konkurencji, która może być "Men" (mężczyźni) lub "Women" (kobiety).
8. `Team`: Trzy literowy kod kraju reprezentowanego przez zawodnika lub drużynę.
9. `Athlete`: Imię i nazwisko zawodnika lub nazwa drużyny.
10. `Results`: Wynik uzyskany w konkurencji pływackiej, zazwyczaj wyrażany w formie czasu.
11. `Rank`: Pozycja zawodnika lub drużyny w konkurencji. Wartości to: 0 - nie wystartował/nie ukończył/zdyskwalifikowany, 1 - złoto, 2 - srebro, 3 - brąz, 4 - brak medalu, 5 - brak danych.

```python
import pandas as pd 

swimming_data = pd.read_csv('/kaggle/input/olympic-swimming-1912-to-2020/Olympic_Swimming_Results_1912to2020.csv', decimal='.', sep=',')
swimming_data.head()
```

```python
swimming_data.to_csv('Olympic_Swimming_Results_1912to2020.csv')
```

```python
swimming_data.head()
```

W pierwszym kroku przeprowadź zmianę nazwy kolumny 'Distance (in meters)' na 'Distance' w ramce danych df. Ta operacja pozwala na bardziej czytelną i zwięzłą reprezentację danych.

```python
swimming_data.rename(columns={'Distance (in meters)': 'Distance'}, inplace=True)
print(swimming_data)
```

Następnie dokonaj zmiany typu danych dla kolumn 'Year' i 'Rank' na typ znakowy (string) w ramce danych df. Jest to przydatne, jeśli te kolumny zawierają informacje, które nie będą poddane działaniom matematycznym, a jedynie operacjom na tekstach.


```python
swimming_data['Year'] = swimming_data['Year'].astype(str)
swimming_data['Rank'] = swimming_data['Rank'].astype(str)
```

```python
swimming_data.info()
```

Wykorzystaj funkcję isna().sum(), aby sprawdzić, czy w ramce danych df istnieją jakiekolwiek brakujące wartości (null). Wynik tej operacji może pomóc w identyfikacji, które kolumny zawierają brakujące dane i w jakiej ilości.

```python
missing_values = swimming_data.isna().sum()
print(missing_values)
```

Kolejnym krokiem będzie usunięcie wierszy, w których występowały brakujące dane (null). Brakujące wiersze zostały zidentyfikowane i przypisane do nowej ramki danych na, a następnie usunięte z oryginalnej ramki df za pomocą metody dropna(inplace=True). To pozwala na zachowanie spójności danych i eliminację brakujących rekordów.

```python
na_rows = swimming_data.loc[swimming_data.isnull().any(axis=1)].copy()
na_rows.shape[0]
```

```python
swimming_data.dropna(inplace=True)
swimming_data
```

Jaki jest ogólny podział danych w ramce swimming_data na konkurencje pływackie (Reley i Indywidualne)? Przedstaw proporcje lub ilości wyników dla obu typów konkurencji.

```python
competition_counts = swimming_data['Relay?'].value_counts()
competition_counts
```

```python
competition_proportions = swimming_data['Relay?'].value_counts(normalize=True)
competition_proportions
```

Jakie są najpopularniejsze konkurencje pływackie (rodzaj stylu i dystans)? Wyświetl te konkurencje wraz z ilością wystąpień w danych.

```python
popular_events = swimming_data.groupby(['Stroke', 'Distance']).size().reset_index(name='Count').sort_values(by='Count', ascending=False)
popular_events.reset_index(drop=True, inplace=True)
popular_events
```

Które drużyny osiągnęły najwięcej sukcesów w konkurencjach pływackich typu "Reley" = 1 (sztafety) ? Zidentyfikuj te drużyny i wyświetl ilość zdobytych medali.

```python
relay_filtered = swimming_data.loc[
(swimming_data['Relay?'] == 1) &
(swimming_data['Rank'] != '0')
]
relay_filtered
```

```python
best_teams = relay_filtered.groupby(['Team', 'Gender']).size().reset_index(name='Rank').sort_values(by='Rank', ascending=False)
best_teams
```

Które drużyny osiągnęły najwięcej sukcesów w konkurencjach pływackich typu "Indywidualne"? Zidentyfikuj te drużyny i wyświetl ilość zdobytych medali.

```python
relay_filtered_indyvidual = swimming_data.loc[
(swimming_data['Relay?'] == 0) &
(swimming_data['Rank'] != '0')
]

relay_filtered_indyvidual
```

```python
best_teams_indyvidual = relay_filtered_indyvidual.groupby(['Team', 'Gender']).size().reset_index(name='Rank').sort_values(by='Rank', ascending=False)
best_teams_indyvidual
```

Które konkurencje pływackie (rodzaj stylu i dystans) były obecne na największej liczbie różnych igrzysk olimpijskich? Przedstaw te konkurencje wraz z ilością wystąpień.

```python
competition_counts = swimming_data.groupby(['Stroke', 'Distance'])['Year'].nunique().reset_index(name='Count').sort_values(by='Count', ascending=False)
competition_counts
```

Czy można zidentyfikować kraje, które osiągnęły najwięcej sukcesów w konkurencjach pływackich (zarówno Indywidualne, jak i Reley)? Przedstaw te kraje wraz z ilościami zdobytych medali.

```python
filtered_data = swimming_data[swimming_data['Rank'] != '0']
best_countries = filtered_data.groupby(['Team', 'Rank']).size().reset_index(name='medal_count').sort_values(by='medal_count', ascending=False)
best_countries
```

Jakie są najczęstsze rodzaje stylów pływackich występujące w danych?

```python
swimming_styles = swimming_data.groupby('Stroke').size().reset_index(name='style_count').sort_values(by='style_count', ascending=False)
swimming_styles
```

Czy można zidentyfikować zmiany w popularności określonych stylów pływackich na przestrzeni lat? Przedstaw ewentualne zmiany w ilości wystąpień każdego stylu na przestrzeni lat.

```python
stroke_trends = swimming_data.groupby(['Stroke', 'Year']).size().reset_index(name='Count')
stroke_trends_pivot = stroke_trends.pivot(index='Year', columns='Stroke', values='Count').fillna(0)
stroke_trends_pivot
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 6))
for stroke in stroke_trends_pivot.columns:
    plt.plot(stroke_trends_pivot.index, stroke_trends_pivot[stroke], label=stroke)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trends in Swimming Styles Over Time')
plt.legend(title='Stroke')
plt.grid(True)
plt.show()
```

Które lata były szczególnie udane dla pływaków danej drużyny (np. najwięcej medali zdobytych w jednym roku)? Wskazuj lata, drużyny i ilość zdobytych medali.

```python
data_filtered = swimming_data[swimming_data['Rank'] != 0]

medals_per_year = data_filtered.groupby(['Team', 'Year']).size().reset_index(name='Medal_Count')
best_team_year = medals_per_year.loc[medals_per_year.groupby('Team')['Medal_Count'].idxmax()]
best_team_year_sorted = best_team_year.sort_values(by='Medal_Count', ascending=False)
best_team_year_sorted
```

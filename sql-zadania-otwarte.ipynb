```python
from google.cloud import bigquery

# Create a "Client" object
client = bigquery.Client()

# Construct a reference to the "stackoverflow" dataset
dataset_ref = client.dataset("stackoverflow", project="bigquery-public-data")

# API request - fetch the dataset
dataset = client.get_dataset(dataset_ref)

# Construct a reference to the "posts_questions" table
table_ref = dataset_ref.table("posts_questions")

# API request - fetch the table
table = client.get_table(table_ref)

# Preview the first five lines of the table
client.list_rows(table, max_results=5).to_dataframe()
```

### Przykład zapytania ograniczającego

Wyświetl tytuły i treści dziesięciu pytań z najwyższym wynikiem (score) w tabeli.

```python
query = """
SELECT title, body, score
FROM `bigquery-public-data.stackoverflow.posts_questions`
ORDER BY score DESC
LIMIT 10
"""
result = client.query(query).to_dataframe()
print(result)
```

Znajdź wszystkie pytania, które mają zaakceptowane odpowiedzi (accepted_answer_id).

```python
query = """
SELECT title, body, score, accepted_answer_id
FROM `bigquery-public-data.stackoverflow.posts_questions`
WHERE accepted_answer_id is not null
ORDER BY score DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Policz liczbę pytań dla każdej wartości liczby odpowiedzi (answer_count).

```python
query = """
SELECT answer_count, COUNT(*) as question_count
FROM `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY answer_count
ORDER BY answer_count
"""

result = client.query(query).to_dataframe()
result
```

Oblicz średni wynik (score) pytań dla każdego unikalnego tagu.

```python
query = """
SELECT tag, AVG(score) as average_score
FROM `bigquery-public-data.stackoverflow.posts_questions`,
UNNEST(SPLIT(tags, '|')) as tag
WHERE tags IS NOT NULL
GROUP BY tag
ORDER BY average_score DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Znajdź użytkowników, którzy zadali najwięcej pytań, i wyświetl ich identyfikatory oraz liczby zadanych pytań.

```python
query = """
SELECT owner_user_id, count(*) as question_count
from `bigquery-public-data.stackoverflow.posts_questions`
WHERE owner_user_id is not null
GROUP BY owner_user_id
ORDER BY question_count DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Policz liczbę pytań dla każdej wartości liczby komentarzy (comment_count).

```python
query = """
SELECT comment_count, COUNT(*) as questions_count
from `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY comment_count
ORDER BY questions_count DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Wyświetl tytuły dziesięciu pytań z największą liczbą ulubionych (favorite_count).

```python
query = """
SELECT title, favorite_count
from `bigquery-public-data.stackoverflow.posts_questions`
ORDER BY favorite_count DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Znajdź wszystkie pytania, które były edytowane przez innych użytkowników (last_editor_user_id).


```python
query = """
SELECT title, owner_user_id, last_editor_user_id
FROM `bigquery-public-data.stackoverflow.posts_questions`
WHERE owner_user_id != last_editor_user_id
ORDER BY last_editor_user_id DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Oblicz średnią liczbę wyświetleń pytań (view_count) dla każdego roku utworzenia pytania (creation_date).

```python
query = """
SELECT EXTRACT(YEAR FROM creation_date) as year, CAST(AVG(view_count) as INT64) as average_view_count
FROM `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY year
ORDER BY year
LIMIT 20
"""

result = client.query(query).to_dataframe()
result
```

Znajdź liczbę pytań zawierających określony tag (np. 'python').

```python
query = """
SELECT tag, COUNT(*) as questions_count
FROM `bigquery-public-data.stackoverflow.posts_questions`,
UNNEST(SPLIT(tags, '|')) as tag
WHERE tag = 'python'
GROUP BY tag
"""

result = client.query(query).to_dataframe()
result
```

Wyświetl tytuły dziesięciu pytań o najdłuższym czasie aktywności (różnica między last_activity_date a creation_date).

```python
query = """
SELECT title, TIMESTAMP_DIFF(last_activity_date, creation_date, DAY) as activity_duration_days
FROM `bigquery-public-data.stackoverflow.posts_questions`
ORDER BY activity_duration_days DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```

Policz liczbę pytań dla każdego typu (post_type_id).

```python
query = """
SELECT post_type_id, COUNT(*) as questions_count
FROM `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY post_type_id
ORDER BY questions_count DESC
LIMIT 10
"""

result = client.query(query).to_dataframe()
result
```
